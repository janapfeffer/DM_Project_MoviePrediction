### Import Libraries

```python
import pandas as pd
import re
import warnings
from sklearn import metrics
import numpy as np
warnings.simplefilter(action='ignore', category=FutureWarning)
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn import linear_model, svm
from sklearn.model_selection import cross_val_score
import time
```

### Supporting functions

```python
## Funktion zur Ausführung von Modellen
def executeModel(model, features_used=False, error=False):
    model.fit(features_train,rating_train)
    y_pred = model.predict(features_test)
    
    train_score=model.score(features_train, rating_train)
    test_score=model.score(features_test, rating_test)
    r2 = metrics.r2_score(rating_test,y_pred)
    
    print("Train score:", train_score)
    print("Test score:", test_score)
    
    if(features_used):
        coeff_used = np.sum(model.coef_!=0)
        print("Number of features used: ", coeff_used)
    
    if(error):
        print("Loss: ", model.loss_)
    
    print("MAE: ", metrics.mean_absolute_error(rating_test, y_pred))
    print("MSE: ", metrics.mean_squared_error(rating_test, y_pred))
    print("RMSE: ", np.sqrt(metrics.mean_squared_error(rating_test, y_pred)))
    print("R^2: ", r2) # sollte hoch sein (je näher an 1, desto besser)
```

## Import Data from Preprocessing

```python
#df_joined = pd.read_csv("regressionPreprocessing_short.csv")
df_joined = pd.read_csv("regressionPreprocessingGeneral.csv")
df_joined = df_joined.fillna(0)
print("Length after import: " + str(len(df_joined)))
print("Features after import: " + str(len(df_joined.columns)))
```

### Remove Features (if needed)

```python
features_to_remove = ['movieId','id','director','+18','18+','actors','productionCompanies','productionCountries','imdbId','spokenLanguages','budget_norm','runtime_norm']
for i in features_to_remove:
    if i in df_joined.columns:
        df_joined = df_joined.drop(columns=i)
print("Features: ", len(df_joined.columns))
df_joined.head(5)
```

## Train and Test Split

```python
from sklearn.model_selection import train_test_split

df_joined_copy = df_joined.copy()
rating = df_joined['rating']
features = df_joined_copy.drop(columns=['rating'])

features_train, features_test, rating_train, rating_test = train_test_split(
    features, rating, test_size=0.1, random_state=42)

print("Train: " + str(len(features_train)) + " Feature Entries and " + str(len(rating_train)) + " Ratings")
print("Test: " + str(len(features_test)) + " Features Entries and " + str(len(rating_test)) + " Ratings")
```

## Correlation Analysis

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# Pearson Correlation (0 no correlation, 1 stronger positive correlation, -1 stronger negative correlation)
plt.figure(figsize=(40,30))
cor = df_joined.corr()
sns.heatmap(cor, annot=True, cmap = plt.cm.Reds)
plt.show()
```

```python
cor_target = abs(cor['rating'])

# selecting highly correlated features
relevant_features = cor_target[cor_target > 0.2]
print(relevant_features)
```

## Feature Selection with XGBRegressor

### Which features are the most important / have the most impact on rating?

```python
from xgboost import XGBRegressor
regressor = XGBRegressor(colsample_bytree= 0.6, gamma= 0.7, max_depth= 4, min_child_weight= 5,
                         subsample = 0.8, objective='reg:squarederror')
regressor.fit(features_train, rating_train)

importances = {}

count = 0
for feature_importance in regressor.feature_importances_:
    if feature_importance > 0.01:
        feature_name = features_train.columns[count]
        importances[feature_name] = feature_importance
    count+=1

for key, value in sorted(importances.items(), key=lambda item: item[1], reverse=True):
    print("%s: %s" % (key, value))
```

### Remove features, which are not important

```python
rating = df_joined['rating']

for i in df_joined.columns:
    if i not in importances:
        df_joined = df_joined.drop(columns=i)

print("Selected features: ", len(df_joined.columns))
df_joined.head(5)
```

## Split again, only with choosen attributes

```python
features = df_joined
# create a train/test split
features_train, features_test, rating_train, rating_test = train_test_split(
    features, rating, test_size=0.1, random_state=42)

print("Train: " + str(len(features_train)) + " Feature Entries and " + str(len(rating_train)) + " Ratings")
print("Test: " + str(len(features_test)) + " Features Entries and " + str(len(rating_test)) + " Ratings")
```

## Linear, Ridge and Lasso Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

print("### Linear Regression ###")
model = LinearRegression()
executeModel(model)

print("\n### Ridge (Alpha = 0.001) ###")
model = Ridge(alpha=0.001)
executeModel(model)

#print("\n### Ridge (Alpha = 10) ###")
#model = Ridge(alpha=10)
#executeModel(model)

#print("\n### Ridge (Alpha = 100) ###")
#model = Ridge(alpha=100)
#executeModel(model)

#print("\n### Bayesian Ridge ###")
#model = BayesianRidge()
#executeModel(model)

print("\n### Lasso Standard (Alpha=1) ###")
model = Lasso()
executeModel(model, features_used=True)

print("\n### Lasso (low Alpha = 0.01) ###")
model = Lasso(alpha=0.01, max_iter=10e5)
executeModel(model, features_used=True)

print("\n### Lasso (very low Alpha = 0.0001) ###")
model = Lasso(alpha=0.0001, max_iter=10e5)
executeModel(model, features_used=True)

print("\n### MLP Regressor (multilayer neuronal network) ###")
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(15,),
            activation='relu', # 'logistic' / 'relu'
            solver='sgd', # 'sgd'/'adam'
            learning_rate='adaptive', # default ‘constant’
            max_iter=10000,
            learning_rate_init=0.01,
            alpha=0.0001) # default 0.0001
executeModel(model, error=True)
```

### Hyperparameter Tuning for Linear Regression

```python
regrOLS = linear_model.LinearRegression()

# Perform 5 fold cross-validation and store the MSE resulted from each fold
scores = cross_val_score(regrOLS,features_train,rating_train,cv=5)

lambdas = [0.0001, 0.001, 0.01, 0.02, 0.05, 0.1, 1, 10]
for l in lambdas:
    print ('Lambda = %.5f'% l)
    # Start time for the 5-fold CV
    start = time.time()
    # Create ridge regression object
    regrRidge = linear_model.Ridge(alpha=l)
    scores = cross_val_score(regrRidge,features_train,rating_train,cv=5)
    end = time.time()
    t = end-start
    print ('Ridge: Least CV error: %.2f and time : %.3f' % (np.min(-scores), t))
    start = time.time()
    # Create lasso object
    regrLasso = linear_model.Lasso(alpha=l)
    scores = cross_val_score(regrLasso,features_train,rating_train,cv=5)
    # Measure and compute time for the 5-fold CV
    end = time.time()
    t = end-start
    print ('Lasso: Least CV error: %.2f and time : %.3f' % (np.min(-scores), t))
    print ('\n')
```

### Hyperparameter Tuning for Random Forest Regressor

```python
# Range of hyperparamters to choose for CV
numTrees = [1, 2, 5, 10, 20, 35, 50, 100, 200]
maxFeatures = [0.25, 0.5, 0.75, 1]
maxDepth = [3, 6, 8, 10, 15, 25]
for n in numTrees:
    for mf in maxFeatures:
        for d in maxDepth:
            print ('Number of trees/estimators = %d, max depth = %d'% (n,d))
            start = time.time()
            # Create Random Forest Regressor object
            randFor = RandomForestRegressor(max_depth=d, random_state=0, n_estimators=n, max_features=mf)
            scores = cross_val_score(randFor,features_train,rating_train,cv=5)
            end = time.time()
            t = end-start
            print ('Random forest regressor: %% of features = %.2f, Least CV error: %.2f and time : %.3f'% (100*mf, np.min(-scores), t))
        print ('\n')
```

### Hyperparameter Tuning for AdaBoost Regressor

```python
# Range of hyperparamters to choose for CV
numTrees = [1, 2, 5, 10, 20, 35, 50, 100, 200]
lossType = ['linear', 'square', 'exponential']
maxDepth = [3, 6, 8, 10, 15, 25]
for l in lossType:
    for n in numTrees:
        for d in maxDepth:
            print ('Number of trees/estimators = %d, max depth = %d'% (n,d))
            start = time.time()
            # Create Boosting Regressor object
            boosting = AdaBoostRegressor(DecisionTreeRegressor(max_depth=d),random_state=0, n_estimators=n, loss=l)
            scores = cross_val_score(boosting,features_train,rating_train,cv=5, n_jobs=1)
            end = time.time()
            t = end-start
            print ('Regressor using boosting: loss type = %s, Least CV error: %.2f and time : %.3f'% (l, np.min(-scores), t))
        print ('\n')
```

### Random Forest Regressor - Trying with the best result

```python
# num Tree = 1; max Features 0.75; max depth 25 --> mehr versuchen

# Create Random Forest Regressor object
randFor = RandomForestRegressor(max_depth=200, random_state=0, n_estimators=1, max_features=0.75)
scores = cross_val_score(randFor,features_train,rating_train,cv=5)
print ('Random forest regressor: %% of features = %.2f, Least CV error: %.2f'% (100*0.75, np.min(-scores)))
executeModel(randFor)
```

### AdaBoost Regressor - Trying with the best result

```python
# exponential (loss), num of trees: 1 (n_estimators), max depth 25 -> mehr versuchen

# Create Boosting Regressor object
boosting = AdaBoostRegressor(DecisionTreeRegressor(max_depth=200),random_state=0, n_estimators=1, loss='linear')
scores = cross_val_score(boosting,features_train,rating_train,cv=5, n_jobs=1)
print ('Regressor using boosting: loss type = %s, Least CV error: %.2f'% ('exponential', np.min(-scores)))
executeModel(boosting)
```
